```
x = np.arange(4)
xx = x.reshape(4,1)
y = np.ones(5)
(xx + y)

print(xx.base is x)
#在上面第二行，將x.reshape(4,1)得到的結果給了xx，這表示xx是從x來的，所以True
x = np.arange(4)
xx = x.reshape(4,1)
y = np.ones(5)
(xx + y)

print(xx.base is y)
```

```
x = np.array([[2,3,4], [5,6,7]])
np.cumsum(x)
#第一個index為2 = 2
#第二個index為2+3 = 5
#第三個index為2+3+4 = 9
#第四個index為2+3+4+5 = 14
#以此類推 。
```
